```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 23, 24, 22, 20, 19, 21]

fig, ax1 = plt.subplots()
ax1.plot(days, temperatures, marker='X', linestyle='--', markerfacecolor="red", color='b')
plt.show()


```

```python
math_scores = [85, 70, 92, 78, 88, 95, 60, 75, 85, 70]
science_scores = [75, 65, 88, 72, 80, 92, 55, 70, 78, 68]

plt.scatter(science_scores, math_scores, c="red", marker="X", label="Students")
plt.title("Scatter graph")
plt.xlabel("Science")
plt.ylabel("Math")
plt.legend()
plt.grid(True)
plt.show()
```

```python
scores = np.array([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6],dtype="int")
plt.hist(scores, bins=5, edgecolor="black", color="yellow")
plt.ylabel("Number of Students")
plt.xlabel("Socres")
plt.title("Histogram")
plt.show()
```

```python
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]

plt.bar(products, sales, color=["green", "red", "blue"], edgecolor="black")
plt.title("Bars")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
products = ['toys', 'food', 'drinks']
sales = [100, 50, 200]

plt.barh(products, sales, color=["green", "red", "blue"], edgecolor="black")
plt.title("Horizontal Bars")
plt.xlabel("Sales")
plt.ylabel("Products")
plt.show()
```

```python
job_types = ['Engineer', 'Manager', 'Analyst', 'Designer', 'Salesperson']
incomes = [80000, 95000, 75000, 65000, 60000]

plt.boxplot(incomes)
plt.ylabel("Income (USD)")
plt.xlabel('Job type')
plt.title("Boxplot")
plt.grid(True)
plt.show()
```

```python
budget_categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Savings']
budget_allocation = [30, 20, 15, 10, 25]

plt.pie(budget_allocation, labels=budget_categories, autopct='%1.1f%%', startangle=140)
plt.title("Monthly Budget Allocation")
plt.show()
```
